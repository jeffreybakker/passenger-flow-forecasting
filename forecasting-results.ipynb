## Metrics

```python
from sklearn import metrics

def generate_metrics(res: pd.DataFrame, y_true: str, y_pred: str):
    # noinspection PyShadowingNames
    def compute(y_true: pd.Series, y_pred: pd.Series):
        y_true = y_true.astype('int')
        y_pred = y_pred.astype('int')

        # Drop results where y_true == 0
        mask = y_true > 0
        y_true = y_true[mask]
        y_pred = y_pred[mask]

        return pd.Series({
            'mae': metrics.mean_absolute_error(y_true, y_pred),
            'mape': metrics.mean_absolute_percentage_error(y_true, y_pred) * 100,
            'rmse': metrics.mean_squared_error(y_true, y_pred, squared=False),
            'corr': y_true.corr(y_pred),
            'r2': metrics.r2_score(y_true, y_pred)
        })

    return res[['model', 'seed', y_true, y_pred]] \
        .reset_index(drop=True) \
        .groupby('model').apply(lambda df: df.groupby('seed').apply(lambda dfa: compute(dfa[y_true], dfa[y_pred])))

metrics_distribution = lambda df: df.apply(lambda col: {'mean': col.mean(), 'std': col.std()})
format_mean = lambda col: col.apply(lambda x: f"${x['mean']:6.2f}$" if col.name not in ['corr', 'r2'] else f"${x['mean']:.4f}$")
format_dist = lambda col: col.apply(lambda x: f"${x['mean']:6.2f} \pm {x['std']:.2f}$" if col.name not in ['corr', 'r2'] else f"${x['mean']:.4f} \pm {x['std']:.4f}$")

generate_metrics(results_no_events, 'passengers', 'step_01').groupby('model').apply(metrics_distribution).apply(format_dist)
```

#### Event conditions samples

```python
from functools import reduce

# The radius of stations around the event to consider during event conditions
EVENT_HOPS = 2

# Compute direct neighbours of every station
neighbours = reduce(
    lambda neighbours, edge: neighbours | {edge[0]: neighbours[edge[0]] + [edge[1]] if edge[0] in neighbours else [edge[1]]},
    data.droplevel(0).index.to_series().unique().tolist(),
    {}
)

# Compute the neighbourhood of every station
neighbourhood = {
    station: [(station, n) for n in neighbours[station]]
    for station in neighbours.keys()
}
for i in range(1, EVENT_HOPS):
    neighbourhood = {
        station: list(set(edges + [(dst, n) for src, dst in edges for n in neighbours[dst] if (dst, n) not in edges]))
        for station, edges in neighbourhood.items()
    }

neighbourhood
```

```python
from datetime import timedelta

# Find all events and note their date and nearest station
event_samples = data[data.event_capacity > 0] \
    .reset_index(drop=False) \
    .rename(columns={'destination': 'station'}) \
    .drop_duplicates(['datetime', 'station']) \
    [['datetime', 'station']]

# Get the neighbourhood of every observation and make a copy for later use
event_samples['neighbourhood'] = event_samples.station.apply(lambda station: neighbourhood[station])
event_sample_list = event_samples.copy()


def select_event_samples(row: pd.Series):
    dt, station = row.name
    neighbourhood = row.neighbourhood

    # Prepare resulting mask
    mask = results_events.assign(mask=False)['mask']

    # Enable all observations in the neighbourhood
    for src, dst in neighbourhood:
        mask[src, dst] = True

    # Swap index levels (needed because duplicate indices)
    mask = mask.swaplevel(0, 2).sort_index()

    # Disable all observations outside of time window (one hour extra than [-4; +2] because of how slicing in pandas works)
    time_start = dt - timedelta(hours=5)
    time_end = dt + timedelta(hours=3)
    mask[:time_start] = False
    mask[time_end:] = False

    # Reset index order
    mask = mask.swaplevel(0, 2).sort_index()

    return mask


event_samples = event_samples \
    .set_index(['datetime', 'station']) \
    .progress_apply(select_event_samples, axis=1) \
    .apply(lambda col: reduce(lambda a, b: a | b, col))

results_no_events[event_samples]
```

```python
normal_samples = event_samples.to_frame('mask')
normal_samples = normal_samples.assign(weekday=normal_samples.apply(lambda row: row.name[2].weekday() < 5, axis=1),
                                       hour=normal_samples.apply(lambda row: row.name[2].hour, axis=1))


def select_normal_samples(df: pd.DataFrame) -> pd.DataFrame:
    normal = df[~df['mask']].sample(frac=1.0, random_state=SEED)
    events = df[df['mask']]

    events['mask'] = False
    normal['mask'].iloc[:events.shape[0]] = True

    return pd.concat([normal, events])


normal_samples = normal_samples \
    .groupby(['origin', 'destination', 'weekday', 'hour']).progress_apply(select_normal_samples) \
    .droplevel([0, 1, 2, 3]) \
    .sort_index()

normal_samples = normal_samples['mask']

print(f"Normal samples: {results_no_events[normal_samples].shape[0]}")
print(f"Event samples: {results_no_events[event_samples].shape[0]}")
print(f"Overlap: {results_no_events[normal_samples & event_samples].shape[0]}")

results_no_events[normal_samples]
```

```python
balanced_event_samples = normal_samples.to_frame('normal').assign(event=event_samples)
balanced_event_samples = balanced_event_samples.assign(
    weekday=balanced_event_samples.apply(lambda row: row.name[2].weekday() < 5, axis=1),
    hour=balanced_event_samples.apply(lambda row: row.name[2].hour, axis=1))


def select_balanced_samples(df: pd.DataFrame) -> pd.DataFrame:
    normal = df[df.normal]
    events = df[df.event].sample(frac=1.0, random_state=SEED)
    neither = df[~(df.normal | df.event)]

    events.event.iloc[normal.shape[0]:] = False

    return pd.concat([normal, events, neither])


balanced_event_samples = balanced_event_samples \
    .groupby(['origin', 'destination', 'weekday', 'hour']).progress_apply(select_balanced_samples) \
    .droplevel([0, 1, 2, 3]) \
    .sort_index()

balanced_event_samples = balanced_event_samples.event

print(f"Normal samples: {results_no_events[normal_samples].shape[0]}")
print(f"Event samples: {results_no_events[balanced_event_samples].shape[0]}")
print(f"Overlap: {results_no_events[normal_samples & balanced_event_samples].shape[0]}")
```

```python
potential_examples = results_events[['passengers', 'model', 'seed', 'step_24']] \
    .rename(columns={'step_24': 'event'}) \
    .assign(no_event=results_no_events.step_24)

# Error of forecast without event-features
potential_examples['error'] = potential_examples.passengers - potential_examples.no_event

# Difference in forecasts between with and without event-features
potential_examples['diff'] = potential_examples.event - potential_examples.no_event

# How much the error is accounted for in the forecast with event-features
potential_examples['percentage'] = potential_examples['diff'] / potential_examples['error']

# Filter out the less interesting cases
potential_examples = potential_examples[
    (potential_examples['error'].abs() > 300)
    & ((potential_examples['event'] - potential_examples['passengers']).abs() < potential_examples['error'].abs())
    & (potential_examples['percentage'] > 0.5)
    & (potential_examples['percentage'] < 1.25)
]

potential_examples['date'] = potential_examples.apply(lambda row: row.name[2].date(), axis=1)
potential_examples = potential_examples.groupby(['origin', 'destination', 'date', 'model']).sum()
potential_examples['percentage'] = potential_examples['diff'] / potential_examples['error']

potential_examples.sort_values('diff', ascending=False).loc[:, :, :, 'mlp']
```

#### Residual diagnostics

```python
from matplotlib.axes import Axes
from matplotlib.lines import Line2D
from matplotlib.ticker import NullLocator


def plot_dist(dist: pd.Series, color, ax: Axes):
    sns.kdeplot(dist,
                bw_adjust=2,
                gridsize=1000,
                color=color,
                ax=ax)

    mean = dist.mean()
    stdev = dist.std()
    coords = np.interp([mean - stdev, mean, mean + stdev],
                       ax.get_lines()[-1].get_path().vertices[:, 0],
                       ax.get_lines()[-1].get_path().vertices[:, 1])
    ax.add_line(Line2D([mean, mean],
                       [0, coords[1]],
                       color=color,
                       linestyle='--'))
    # ax.add_line(Line2D([mean - stdev, mean - stdev],
    #                    [0, coords[0]]))
    # ax.add_line(Line2D([mean + stdev, mean + stdev],
    #                    [0, coords[2]]))


error_dist = results_events \
    .assign(error=results_events.step_01 - results_events.passengers) \
    .drop(columns=filter(lambda x: x.startswith('step_'), results_events.columns)) \
    .drop(columns='passengers')

error_dist_event = results_events[balanced_event_samples]
error_dist_event = error_dist_event \
    .assign(error=error_dist_event.step_01 - error_dist_event.passengers) \
    .drop(columns=filter(lambda x: x.startswith('step_'), error_dist_event.columns)) \
    .drop(columns='passengers')

fig, axes = plt.subplots(2, 4, figsize=(14, 3))
max_y = 0.0
for i, model in enumerate(tqdm([m for m in model_order if m is not None])):
    ax = axes[i // 4, i % 4]

    plot_dist(error_dist[error_dist.model == model].error,
              colormaps['tab20'].colors[i * 2],
              ax)
    plot_dist(error_dist_event[error_dist_event.model == model].error,
              colormaps['tab20'].colors[i * 2 + 1],
              ax)

    style_plot(ax)

    ax.set_xlim(-150, 150)
    ax.set_xlabel(None)

    ax.yaxis.set_major_formatter(NullFormatter())
    ax.set_ylabel(None)

    ax.spines['left'].set_visible(False)

    max_y = max(max_y, ax.get_ylim()[1])

for i, model in enumerate(m for m in model_order if m is not None):
    ax = axes[i // 4, i % 4]
    ax.set_ylim(0, max_y)

fig.legend(handles=legend_elements,
           ncols=7,
           bbox_to_anchor=(0.12, -0.05, 0.8, 0.1),
           mode='expand',
           frameon=False)
fig.legend(title='Subset',
           handles=[Patch(facecolor=(0, 0, 0, 0.5), label='Overall'),
                    Patch(facecolor=(0, 0, 0, 0.1), label='Event conditions')],
           bbox_to_anchor=(0.875, 0.375))

ax = axes[-1, -1]
ax.grid(False)
ax.xaxis.set_major_locator(NullLocator())
ax.yaxis.set_major_locator(NullLocator())
ax.patch.set_alpha(0.0)

plt.savefig('figures/metrics-bias.png', dpi=300, bbox_inches='tight', transparent=True)
```
